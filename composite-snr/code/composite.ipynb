## 将两个表格合并

```python
def normalize_g_name(g_name):
    parts = g_name[1:].split("+") if "+" in g_name else g_name[1:].split("-")
    l_part = parts[0].zfill(5)  # 填充到5位
    b_part = parts[1].zfill(4)
    return f"G{l_part}{'+' if '+' in g_name else '-'}{b_part}"

# 对两个文件中的G name进行标准化处理G
# 在两个表格中添加一列source from，用于标记来源
df_composite["source_from"] = "snrcat"
df_green_com["source_from"] = "green"
#df_composite["normalized_G"] = df_composite["G"].apply(normalize_g_name)
df_green_com["normalized_name"] = df_green_com["name"].apply(normalize_g_name)

# 合并两个DataFrame，基于规范化的G name 进行匹配
df_merge = pd.merge(df_composite, df_green_com, left_on="G", right_on="normalized_name", how="outer")
print(df_merge.shape[0])


# 保存df_merge
df_merge.to_csv("../cat/composite_merged.csv", index=False)
```
